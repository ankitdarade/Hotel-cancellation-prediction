```python
from sklearn.preprocessing import StandardScaler
df_encoded = pd.get_dummies(df, columns=['hotel', 'meal', 'market_segment', 'distribution_channel',
                                         'reserved_room_type', 'assigned_room_type', 'deposit_type',
                                         'customer_type'], drop_first=True)

# Standardize numerical features
features_to_scale = ['lead_time', 'total_stay', 'adults', 'children', 'babies', 'adr', 'days_in_waiting_list']
scaler = StandardScaler()
df_encoded[features_to_scale] = scaler.fit_transform(df_encoded[features_to_scale])
```

```python

# Define the list of features you want to check for outliers
features_to_check=['lead_time', 'total_stay', 'adults', 'children', 'babies', 'adr', 'days_in_waiting_list']

def identify_outliers_iqr(df, feature):
    Q1=df[feature].quantile(0.25)
    Q3=df[feature].quantile(0.75)
    IQR=Q3 - Q1
    lower_bound=Q1 - 1.5 * IQR
    upper_bound=Q3 + 1.5 * IQR
    return df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]

outlier_summary = {}

for feature in features_to_check:
    outliers=identify_outliers_iqr(df_encoded, feature)
    outlier_summary[feature]=outliers.shape[0]  

for feature, num_outliers in outlier_summary.items():
    print(f"Number of outliers in {feature}: {num_outliers}")
```

```python

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_check):
    plt.subplot(2, 4, i + 1)
    sns.boxplot(df_encoded[feature])
    plt.title(f'Box Plot of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Removing outliers 
def remove_outliers_iqr(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
df_cleaned = df_encoded.copy()
for feature in features_to_check:
    df_cleaned = remove_outliers_iqr(df_cleaned, feature)

print(f"Shape of the dataset after removing outliers: {df_cleaned.shape}")
```

```python
def impute_outliers(df, feature):
    Q1=df[feature].quantile(0.25)
    Q3=df[feature].quantile(0.75)
    IQR=Q3 - Q1
    lower_bound=Q1 - 1.5 * IQR
    upper_bound=Q3 + 1.5 * IQR
    median=df[feature].median()
    df[feature]=np.where((df[feature] < lower_bound) | (df[feature] > upper_bound), median, df[feature])
    return df
df_imputed=df_encoded.copy()
for feature in features_to_check:
    df_imputed=impute_outliers(df_imputed, feature)

print(f"Shape of the dataset after imputing outliers: {df_imputed.shape}")
```

```python
#Hypothesis 1: Customers Booking in Advance and Cancellations
#For this hypothesis, we can use a Chi-Square test to compare cancellation rates for bookings made more than 6 months (180 days) in advance.

from scipy.stats import chi2_contingency

# Define a new column for booking lead time > 6 months (180 days)
df['long_lead_time']=np.where(df['lead_time'] > 180, 1, 0)

# Create a contingency table for cancellations and long lead times
contingency_table=pd.crosstab(df['is_canceled'], df['long_lead_time'])

# Perform the Chi-Square test
chi2, p, dof, expected=chi2_contingency(contingency_table)

# Output the test results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the result
alpha=0.05  # 95% confidence level
if p < alpha:
    print("Reject the null hypothesis: Customers booking more than 6 months in advance are more likely to cancel.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that long lead time affects cancellations.")

```

```python
# Hypothesis 2: Weekday vs. Weekend Bookings ADR
```

```python
from scipy.stats import ttest_ind

# Create groups based on the stays in weekend vs. weekday nights
weekday_bookings=df[df['stays_in_week_nights'] > 0]['adr']
weekend_bookings=df[df['stays_in_weekend_nights'] > 0]['adr']

# Perform an independent t-test
t_stat, p_value=ttest_ind(weekday_bookings, weekend_bookings, equal_var=False)

# Output the test results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05  # 95% confidence level
if p_value < alpha:
    print("Reject the null hypothesis: Weekday bookings have a higher ADR than weekend bookings.")
else:
    print("Fail to reject the null hypothesis: No significant difference in ADR between weekday and weekend bookings.")
```
